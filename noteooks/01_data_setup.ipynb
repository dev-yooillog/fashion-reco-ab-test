```python
import pandas as pd
import numpy as np
import os
```

```python
transactions = pd.read_csv(
    "../data/raw/transactions_train.csv",
    parse_dates=["t_dat"],
    dtype={
        "article_id": str,
        "customer_id": str,
        "price": float,
        "sales_channel_id": int,
    },
)
transactions.shape
```

```python
transactions.head()
```

```python
transactions = transactions[transactions["t_dat"] >= "2020-03-01"].copy()
transactions.shape
```

```python
np.random.seed(42)
sampled_users = (
    transactions["customer_id"]
    .drop_duplicates()
    .sample(50000, random_state=42)
)

df = transactions[transactions["customer_id"].isin(sampled_users)].copy()
df.shape
```

```python
df.to_parquet("../data/processed/transactions_sample.parquet", index=False)
```

```python
articles = pd.read_csv(
    "../data/raw/articles.csv",
    dtype={"article_id": str},
)

customers = pd.read_csv(
    "../data/raw/customers.csv",
    dtype={"customer_id": str},
)

articles.to_parquet("../data/processed/articles.parquet", index=False)
customers.to_parquet("../data/processed/customers.parquet", index=False)
```
